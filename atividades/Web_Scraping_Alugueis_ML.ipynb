# **WEB SCRAPING**

 Extração de dados do site [Mercado Livre Alugueis](https://imoveis.mercadolivre.com.br/casas/aluguel/). Este exercicio faz parte da formação em Ciência de Dados da Awari School.

```
#Importações para esse exercicio
import requests
import re
from bs4 import BeautifulSoup
import json
import pandas as pd
import numpy as np

```

```
url = 'https://imoveis.mercadolivre.com.br/casas/aluguel/sao-paulo/sao-paulo-zona-sul/'
```

```
#requisição para trabalhar com os dados
requests.get(url)
```

```
#extraindo o conteudo da pagina
conteudo = requests.get(url).content
```

```
conteudo_estruturado = BeautifulSoup(conteudo)
```

```
conteudo_estruturado.text
```

```
#Extraindo o Preço dos imoveis
conteudo_estruturado.find('span', class_='price__fraction').text
```

```
# Colocar todos os preçoes em uma lista
lista_precos = conteudo_estruturado.find_all('span', class_='price__fraction')
```

```
lista_precos[:5]
```

```
# Criando um loop para extrair apenas o valor do aluguel
valor_aluguel = []
for preco in lista_precos:
  valor_aluguel.append(preco.text)
```

```
valor_aluguel[:5]
```

```
# Extraindo a metragem e quarto
conteudo_estruturado.find('div', class_="item__attrs")
```

```
lista_metragem_sujo = conteudo_estruturado.find_all('div', class_='item__attrs')
```

```
#extraindo apenas a metragem
re_area = '> (.*) m²'
```

```
lista_metragem_sujo[0]
```

```
re.findall(re_area, str(lista_metragem_sujo[0]))
```

```
metragem = []
for elemento in lista_metragem_sujo:
  area = re.findall(re_area, str(elemento))
  if area==[]:
    # not a number
    area=np.nan
    metragem.append(area)
  else:
    metragem.append(area[0])
```

```
metragem[:5]
```

```
# Extraindo a quantidade quartos
re_quarto ='\| (.*) quarto'
```

```
quartos = []
for elemento in lista_metragem_sujo:
  area = re.findall(re_quarto, str(elemento))
  if area==[]:
    # not a number
    area=np.nan
    quartos.append(area)
  else:
    quartos.append(area[0])
```

```
quartos[:5]
```

```
# Extraindo o Endereço
conteudo_estruturado.find('div', class_="item__title")
```

```
lista_end_sujo = conteudo_estruturado.find_all('div', class_="item__title")
```

```
lista_end_sujo
```

```
lista_end_sujo = conteudo_estruturado.find_all('div', class_="item__title")

endereco = []
for end in lista_end_sujo:
  endereco.append(end.text)
```

```
endereco[:5]
```

```
# Criando o dicionario
dados = {'area': metragem,
'preco': valor_aluguel,
'quartos': quartos, 'endereço':endereco}
```

```
# Transformando em Data Frame
df = pd.DataFrame(dados)
```

```
# Exportando o o Dataframe para um arquivo CSV
df.to_csv('alugueis_mercado_livre.csv')
```
